```python
def to_datetime_series(series):
    """Best-effort datetime parsing; coerces invalid to NaT."""
    return pd.to_datetime(series, errors="coerce", infer_datetime_format=False)

def summarize_missing(df, name):
    miss = df.isna().mean().sort_values(ascending=False)
    print(f"\nMissingness summary for {name} (top 20):")
    print((miss * 100).round(2).head(20).astype(str) + "%")

def iqr_outlier_bounds(s):
    """Return lower/upper bounds using IQR rule."""
    q1, q3 = np.nanpercentile(s, [25, 75])
    iqr = q3 - q1
    return q1 - 1.5 * iqr, q3 + 1.5 * iqr

def robust_smape(y_true, y_pred):
    denom = (np.abs(y_true) + np.abs(y_pred))
    # avoid divide-by-zero: where denom==0, set term to 0
    out = np.where(denom == 0, 0, 2.0 * np.abs(y_pred - y_true) / denom)
    return np.mean(out) * 100.0

def safe_mape(y_true, y_pred, eps=1e-6):
    denom = np.maximum(np.abs(y_true), eps)
    return np.mean(np.abs((y_true - y_pred) / denom)) * 100.0

def print_date_range(df, ts_col, name):
    print(f"{name} time range: {df[ts_col].min()}  ->  {df[ts_col].max()}  (n={len(df)})")
```

```python
weather = dfweather.copy()
energy = dfenergy.copy()
```

```python
weather.shape, energy.shape
```

```python
energy["timestamp"] = pd.to_datetime(energy["timestamp"], format='%Y-%m-%d %H:%M:%S', errors='coerce')
weather["timestamp"] = pd.to_datetime(
    weather['Date'].astype(str).str.strip() + ' ' + weather['Hour'],
    format='%d/%m/%Y %H:%M', errors='coerce'
)
```

```python
energy = (energy
          .dropna(subset=['timestamp'])
          .drop_duplicates(subset=['timestamp'])
          .sort_values('timestamp'))

energy.shape
```

```python
weather = (weather
           .drop(columns=[c for c in ['Date', 'Hour'] if c in weather.columns])
           .dropna(subset=['timestamp'])
           .drop_duplicates(subset=['timestamp'])
           .sort_values('timestamp'))

weather.shape
```

```python
print_date_range(weather, "timestamp", "Weather")
print_date_range(energy, "timestamp", "Energy")
```

```python
energy["timestamp"].duplicated().sum()
```

```python
summarize_missing(weather, "weather")
summarize_missing(energy,  "energy")
```

# 3. Align by timestamp

```python
merged = pd.merge(
    weather,
    energy[["timestamp", "is_weekend", "season", "is_holiday", "total_consumption_mw"]],
    left_on="timestamp",
    right_on="timestamp",
    how="inner",
    validate="one_to_one"
)
merged
```

```python
weather.shape, energy.shape, merged.shape
```

```python
merged = merged.sort_values("timestamp").reset_index(drop=True)

# Sanity checks
print("\nMerged shape:", merged.shape)
print_date_range(merged, "timestamp", "Merged")
```
